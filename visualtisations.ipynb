## Explore HF and CS2 output

```julia
using MAT

hf = matread("sols_all_20260409_140319.mat")
cs2 = matread("sols_all_20260409_140844.mat")

# typeof(hf)
# keys(hf)
# hf

# typeof(cs2)
# keys(cs2)
# cs2
```

```julia
using Plots

u_vars = [
    "dO2", "dtCO2", "dtNO3", "dtSO4", "dtPO4", "dtNH4", "dtH2S", "dFeII",
    "dMnII", "dCH4", "dalk", "dCa", "pfoc", "psoc", "pFeOH3", "pMnO2",
    "pcalcite", "paragonite", "pFeS", "pFeS2", "pS0", "pFeOH3_PO4"
]

u_var_index = Dict(name => i for (i, name) in enumerate(u_vars))
for name in u_vars
    u_var_index[lowercase(name)] = u_var_index[name]
end

function plot_u_comparison(varname; hf_data=hf, cs2_data=cs2)
    name = String(varname)
    idx = get(u_var_index, name, get(u_var_index, lowercase(name), nothing))
    idx === nothing && error("Unknown variable '$varname'. Choose one of: $(join(u_vars, ", "))")

    uhf  = Array(hf_data["u"][idx, :, :])
    ucs2 = Array(cs2_data["u"][idx, :, :])

    size(uhf) == size(ucs2) || error("HF and CS2 have different shapes for $name")

    diff = ucs2 .- uhf

    shared_clims = extrema(vcat(vec(uhf), vec(ucs2)))
    dmax = maximum(abs, diff)
    diff_clims = (-dmax, dmax)

    x = haskey(hf_data, "t") ? vec(hf_data["t"]) : 1:size(uhf, 2)
    y = haskey(hf_data, "z") ? vec(hf_data["z"]) : 1:size(uhf, 1)

    p1 = heatmap(x, y, uhf;
        title="HF: $name", xlabel="Time", ylabel="Depth",
        yflip=true, clims=shared_clims, colorbar=true)

    p2 = heatmap(x, y, ucs2;
        title="CS2: $name", xlabel="Time", ylabel="Depth",
        yflip=true, clims=shared_clims, colorbar=true)

    p3 = heatmap(x, y, diff;
        title="CS2 - HF: $name", xlabel="Time", ylabel="Depth",
        yflip=true, clims=diff_clims, c=cgrad([:blue, :white, :red]), colorbar=true)

    plot(p1, p2, p3; layout=(1, 3), size=(1200, 350))
end

# Example:
# plot_u_comparison("dO2")
```

```julia
display(plot_u_comparison("dO2"))
display(plot_u_comparison("dtCO2"))
display(plot_u_comparison("dtNO3"))
display(plot_u_comparison("dtSO4"))
display(plot_u_comparison("dtPO4"))
display(plot_u_comparison("dtNH4"))
display(plot_u_comparison("dtH2S"))
display(plot_u_comparison("dFeII"))
display(plot_u_comparison("dMnII"))
display(plot_u_comparison("dCH4"))
display(plot_u_comparison("dalk"))
display(plot_u_comparison("dCa"))
display(plot_u_comparison("pfoc"))
display(plot_u_comparison("psoc"))
display(plot_u_comparison("pFeOH3"))
display(plot_u_comparison("pMnO2"))
display(plot_u_comparison("pcalcite"))
display(plot_u_comparison("paragonite"))
display(plot_u_comparison("pFeS"))
display(plot_u_comparison("pFeS2"))
display(plot_u_comparison("pS0"))
display(plot_u_comparison("pFeOH3_PO4"))
```
